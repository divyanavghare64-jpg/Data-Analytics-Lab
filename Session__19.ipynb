```python
# Q2)
import pandas as pd

df = pd.read_csv('/content/cardekho.csv')

df.head(10)

```

```python
# Q3)
# Missing values
print(df.isnull().sum())

# Duplicate rows
print("Duplicate Rows:", df.duplicated().sum())

# Remove duplicate rows
df = df.drop_duplicates()

print("New Shape:", df.shape)
```

```python
# Q4)
df.describe()
```

```python
# Q5)
import matplotlib.pyplot as plt

df.hist(figsize=(12,8), bins=20)
plt.tight_layout()
plt.show()
```

```python
# Q6)
import seaborn as sns
import matplotlib.pyplot as plt

categorical_columns = ['fuel', 'seller_type', 'transmission', 'owner']

for col in categorical_columns:
    plt.figure(figsize=(6,4))
    sns.countplot(x=df[col])
    plt.title(f"Count Plot of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Q7)
numeric_df = df.select_dtypes(include=['int64','float64'])

plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Q8)
X = df.drop('selling_price', axis=1)

y = df['selling_price']

print("Independent Features:")
print(X.columns)

print("\nTarget Variable:")
print(y.name)
```

```python
# Q9)
encoded_df = pd.get_dummies(df, columns=['fuel','seller_type','transmission','owner'])

encoded_df.head()
```

```python
# Q10)
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

num_cols = ['year','km_driven','selling_price','seats']

encoded_df[num_cols] = scaler.fit_transform(encoded_df[num_cols])

encoded_df.head()
```
