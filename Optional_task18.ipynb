```python
# Q1. Import Required Libraries

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
# Q2. Load Dataset

df = pd.read_csv("/content/insurance.csv")

print("First 5 Rows:")
print(df.head())
```

```python
# Q3. Dataset Information

print("Shape:", df.shape)
print(df.info())
print(df.describe())
```

```python
# Q4. Check Missing Values

print(df.isnull().sum())
```

```python
 # Q5. Check Duplicate Rows

print("Duplicate Rows:", df.duplicated().sum())
```

```python
# Q6. Remove Duplicate Rows

df = df.drop_duplicates()

print("New Shape:", df.shape)
```

```python
# Q7. Check Invalid Values

print((df[['age','bmi','children','expenses']] < 0).sum())
```

```python
print(df.columns)
```

```python
# Q8. Encode Categorical Columns

le = LabelEncoder()

df['sex'] = le.fit_transform(df['sex'])
df['smoker'] = le.fit_transform(df['smoker'])
df['region'] = le.fit_transform(df['region'])

print(df.head())
```

```python
# Q9. Feature Scaling

scaler = StandardScaler()

cols = ['age','bmi','children','expenses']

df[cols] = scaler.fit_transform(df[cols])

print(df.head())
```

```python
# Q10. Outlier Detection & Removal

Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)

IQR = Q3 - Q1

df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]

print("Shape After Removing Outliers:", df.shape)
```

```python
# Q11. Save Cleaned Dataset

df.to_csv("insurance_cleaned.csv", index=False)

print("Dataset Saved Successfully.")
```
