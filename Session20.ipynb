```python
# Q1. Preparing Features

import pandas as pd

df = pd.read_csv("ford_car_dataset.csv")

X = df.drop("price", axis=1)
Y = df["price"]

print("Shape of X:", X.shape)
print("Shape of Y:", Y.shape)
```

```python
# Q2. One-Hot Encoding

cat_cols = X.select_dtypes(include=["object"]).columns
print(cat_cols)

X = pd.get_dummies(X, columns=cat_cols).astype(int)

print(X.head())
```

```python
# Q3. Feature Scaling

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

num_cols = ["year", "mileage", "tax", "mpg", "engineSize"]

X[num_cols] = scaler.fit_transform(X[num_cols])

print(X.head())
```

```python
# Q4. Train-Test Split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, Y,
    test_size=0.33,
    random_state=42
)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python

# Q5. Linear Regression

from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)

print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
```

```python
# Q6. Making Predictions

y_pred = model.predict(X_test)

print("First 10 Predicted Values:")
print(y_pred[:10])

print("\nFirst 10 Actual Values:")
print(y_test.iloc[:10].values)
```

```python
# Q7. Model Evaluation using R² Score

from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)

print("R² Score:", r2)
```

Interpretation : -
The R² score shows how well the Linear Regression model predicts the car prices. A value closer to 1 means the model performs very well, while a value closer to 0 means the performance is poor.

```python
# Q8. Saving the Model

import joblib

joblib.dump(model, "LR_ford_car.pkl")
joblib.dump(scaler, "scaler.pkl")
joblib.dump(X.columns.tolist(), "columns.pkl")

print("Model and preprocessing objects saved successfully.")
```
