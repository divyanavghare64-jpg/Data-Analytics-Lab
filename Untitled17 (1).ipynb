```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('heart.csv')
df.head(10)
```

```python
print("Shape of the dataset")
print(df.shape)
```

```python
print("Basic information")
df.info()
```

```python
df.isnull().sum()
```

```python
print("Duplicate Rows")
duplicates = df.duplicated().sum()
print("Number of duplicated rows:",duplicates)

df = df.drop_duplicates()

print("New shape of the dataset:", df.shape)
```

```python
print(df.columns)
```

```python
cholesterol_zoro = (df['Cholesterol'] ==0).sum()
print("Rows with Cholesterol = 0:",cholesterol_zoro)

restingbp_zoro = (df['RestingBP'] ==0).sum()
print("Rows with RestingBP = 0:",restingbp_zoro)

```

```python
# Before Cleaning
print("Before Cleaning:\n")
print(df[['Cholesterol', 'RestingBP']].describe())

# Mean values (excluding zeros)
chol_mean = df[df['Cholesterol'] != 0]['Cholesterol'].mean()
bp_mean = df[df['RestingBP'] != 0]['RestingBP'].mean()

# Replace 0 with mean
df['Cholesterol'] = df['Cholesterol'].replace(0, chol_mean)
df['RestingBP'] = df['RestingBP'].replace(0, bp_mean)

# Round to 2 decimal places
df['Cholesterol'] = df['Cholesterol'].round(2)
df['RestingBP'] = df['RestingBP'].round(2)

# After Cleaning
print("\nAfter Cleaning:\n")
print(df[['Cholesterol', 'RestingBP']].describe())
```

```python
import matplotlib.pyplot as plt

# Function to plot histogram
def plot_histogram(data, column, color):
    plt.hist(data[column], bins=20, color=color, edgecolor='black')
    plt.title(column)
    plt.xlabel(column)
    plt.ylabel("Frequency")

# Create 2x2 subplots
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plot_histogram(df, 'Age', 'skyblue')

plt.subplot(2, 2, 2)
plot_histogram(df, 'RestingBP', 'lightgreen')

plt.subplot(2, 2, 3)
plot_histogram(df, 'Cholesterol', 'orange')

plt.subplot(2, 2, 4)
plot_histogram(df, 'MaxHR', 'pink')

plt.tight_layout()
plt.show()
```

```python
# Identify numerical and categorical columns

numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
categorical_columns = df.select_dtypes(include=['object']).columns

print("Numerical Columns:")
print(list(numerical_columns))

print("\nCategorical Columns:")
print(list(categorical_columns))
```

```python
# Perform One-Hot Encoding
df_encoded = pd.get_dummies(df, drop_first=False)

# Print shape of encoded dataframe
print("Shape of Encoded DataFrame:", df_encoded.shape)

# Print first 5 rows
print("\nFirst 5 Rows of Encoded DataFrame:")
print(df_encoded.head())
```

```python
# Print final shape of encoded dataframe
print("Final Shape of df_encoded:", df_encoded.shape)

# Print all column names
print("\nColumn Names in Final DataFrame:")
print(df_encoded.columns.tolist())
```

```python
# Summary

# 1. I found invalid values where Cholesterol = 0 and RestingBP = 0.
#    I replaced these values with the mean of the respective columns
#    (excluding zero values) and rounded the values to 2 decimal places.

# 2. It is important to handle invalid values because they can affect
#    data analysis and reduce the accuracy of machine learning models.

# 3. One-Hot Encoding converts categorical data into numerical columns
#    so that machine learning algorithms can process the data correctly.

# 4. After cleaning and encoding, the dataset became more consistent,
#    free from invalid values, and ready for further analysis and model building.

```
